```python
# Tagle, Marc Neil V. (M001)

import pandas as pd
import hvplot.pandas as hv
from IPython.display import display
```

```python
# Creating DataFrame from a CSV file.
df = pd.read_csv('sleep_health_and_lifestyle.csv')
```

```python
# 1. Display the records of people categorized as Overweight.
df[df['BMI Category'] == 'Overweight']
```

```python
# 2. Display the records of people with a Sleep Disorder.
df[df['Sleep Disorder'] != 'None']
```

```python
# 3. Display records of people who sleep less than 6 hours per night.
df[df['Sleep Duration'] < 6]
```

```python
# 4. Display the records of people with both high BMI (Obese) and high Blood Pressure (greater than or equal 140/90).

df_copy = df.copy()
df_copy[['Systolic','Diastolic']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)
high_bmi_bp = df_copy[(df_copy['BMI Category'].isin(['Overweight','Obese'])) & (df_copy['Systolic'] >= 140) & (df_copy['Diastolic'] >= 90)].drop(columns=['Systolic','Diastolic'])
high_bmi_bp_df = pd.DataFrame(high_bmi_bp)
display(high_bmi_bp_df)
```

```python
# 5.pd. Display the average Sleep Duration per BMI Category.

sleep_by_bmi = df.groupby(by='BMI Category')[['Sleep Duration']].mean()
display(sleep_by_bmi)
```

```python
# 5.hv. How does Sleep Duration vary by BMI Category? (Line Chart)

def line_hook(plot, element):
    plot.handles["xaxis"].axis_label_text_font = "Arial"
    plot.handles['xaxis'].axis_label_text_font_style = 'bold'
    plot.handles['xaxis'].axis_label_text_color = 'blue'
    plot.handles["yaxis"].axis_label_text_font = "Arial"
    plot.handles['yaxis'].axis_label_text_font_style = 'bold'
    plot.handles['yaxis'].axis_label_text_color = 'blue'
    plot.handles["xaxis"].major_label_text_font = 'Cambria Math'
    plot.handles["yaxis"].major_label_text_font = 'Cambria Math'

sleep_by_bmi.hvplot.line(title='Average Sleep Duration by BMI Category',
                         xlabel='BMI Category', ylabel='Sleep Duration (hours)',
                         width=800, height=500,
                         line_color='red', line_width=3,
                         fontsize={'xticks':11,'yticks':11,'xlabel':12,'ylabel':12,'title':16}
).opts(hooks=[line_hook])
```

```python
# 6.pd. Display the average Stress level per Occupation, sorted in descending order.

stress_by_occupation = df.groupby(by='Occupation')[['Stress Level']].mean().sort_values(by='Stress Level', ascending=False)
display(stress_by_occupation)
```

```python
# 6.hv. How does Average Stress Level vary by Occupation? (Horizontal Bar Chart)

def barh_hook(plot, element):
    plot.handles["xaxis"].axis_label_text_font = "Arial"
    plot.handles['xaxis'].axis_label_text_font_style = 'bold'
    plot.handles['xaxis'].axis_label_text_color = 'maroon'
    plot.handles["yaxis"].axis_label_text_font = "Arial"
    plot.handles['yaxis'].axis_label_text_font_style = 'bold'
    plot.handles['yaxis'].axis_label_text_color = 'maroon'
    plot.handles["xaxis"].major_label_text_font = 'Helvetica'
    plot.handles["yaxis"].major_label_text_font = 'Helvetica'
    
stress_by_occupation.hvplot.barh(title='Average Stress Level by Occupation',
                                 x='Occupation', y="Stress Level", 
                                 color='lightcoral', line_color=None
).opts(hooks=[barh_hook])
```

```python
# 7.pd. Display the total number of people diagnosed with Sleep Apnea, sorted by Age in ascending order.

sleep_apnea_df = df[df['Sleep Disorder'] == 'Sleep Apnea'].groupby(by='Age').size().reset_index(name='Sleep Apnea Cases').set_index(['Age','Sleep Apnea Cases'])
display(sleep_apnea_df)
```

```python
# 7.hv. How do Total Sleep Apnea Cases vary by Age? (Vertical Bar Chart)

def barv_hook(plot, element):
    plot.handles["xaxis"].axis_label_text_font = "Arial"
    plot.handles['xaxis'].axis_label_text_font_style = 'bold'
    plot.handles['xaxis'].axis_label_text_color = 'darkmagenta'
    plot.handles["yaxis"].axis_label_text_font = "Arial"
    plot.handles['yaxis'].axis_label_text_font_style = 'bold'
    plot.handles['yaxis'].axis_label_text_color = 'darkmagenta'
    plot.handles["xaxis"].major_label_text_font = 'Verdana'
    plot.handles["xaxis"].major_label_text_font_style = "bold"
    plot.handles["yaxis"].major_label_text_font = 'Verdana'
    plot.handles["yaxis"].major_label_text_font_style = "bold"

sleep_apnea_df.hvplot.bar(title="Total Sleep Apnea Cases by Age",
                          x='Age',
                          y='Sleep Apnea Cases',
                          color='m',
                          line_color=None
).opts(hooks=[barv_hook])
```

```python
# 8. Display the total number of people with each Sleep Disorder per BMI Category.

disorder_by_bmi = df.groupby(by=['BMI Category', 'Sleep Disorder']).size().unstack()
display(disorder_by_bmi)
```

```python
# 8.hv. How do Sleep Disorder Cases vary by BMI Category? (Stacked Bar Chart)

def sbar_hook(plot, element):
    plot.handles["xaxis"].axis_label_text_font = "Arial"
    plot.handles['xaxis'].axis_label_text_font_style = 'bold'
    plot.handles['xaxis'].axis_label_text_color = 'darkslategray'
    plot.handles["yaxis"].axis_label_text_font = "Arial"
    plot.handles['yaxis'].axis_label_text_font_style = 'bold'
    plot.handles['yaxis'].axis_label_text_color = 'darkslategray'
    plot.handles["xaxis"].major_label_text_font = 'Comic Sans MS'
    plot.handles["xaxis"].major_label_text_font_style = "bold"
    plot.handles["yaxis"].major_label_text_font = 'Comic Sans MS'
    plot.handles["yaxis"].major_label_text_font_style = "bold"

disorder_by_bmi.hvplot.bar(
    stacked=True,
    title='Total Sleep Disorder Cases per BMI Category',
    xlabel='BMI Category', ylabel='Number of Cases',
    cmap='paired'
).opts(hooks=[sbar_hook])
```

```python
# 9.pd. Display the average Heart Rate per BMI Category, sorted in descending order.

hrate_by_bmi = df.groupby(by='BMI Category')[['Heart Rate']].mean().sort_values(by='Heart Rate', ascending=False)
display(hrate_by_bmi)
```

```python
# 9.hv. How do Average Heart Rates differ by BMI Category? (Vertical Bar Chart)

hrate_by_bmi.hvplot.bar(title='Average Heart Rate per BMI Category',
                        xlabel='BMI Category', ylabel='Average Heart Rate', 
                        color='mediumpurple', line_color=None,
                        rot=45
).opts(hooks=[barv_hook])
```

```python
# 10. Display the top 5 Occupations with the highest average Sleep Duration.

df.groupby(by='Occupation')[['Sleep Duration']].mean().sort_values(by='Sleep Duration', ascending=False).head()
```

```python
# 11. Display the top 3 Age Groups with the lowest average Daily Steps.

df.groupby(by='Age')[['Daily Steps']].mean().round(0).astype(int).sort_values(by='Daily Steps', ascending=True).head(3)
```

```python
# 12. Display the age group with the highest average Stress Level.

max_stress_age = df.groupby(by='Age')['Stress Level'].mean().idxmax()
max_stress_value = df.groupby(by='Age')['Stress Level'].mean().max()

age_max_stress = pd.DataFrame({'Age':[max_stress_age], 'Stress Level':[max_stress_value]})
age_max_stress.reset_index(drop=True, inplace=True)
display(age_max_stress)
```
